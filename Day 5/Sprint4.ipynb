```python
import mysql.connector
from mysql.connector import Error
import matplotlib.pyplot as plt

def create_db_connection(host_name, user_name, user_password,db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")
    return connection

def execute_get_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        records = cursor.fetchall()
        print("Query successful")
        return records
    except Error as err:
        print(f"Error: '{err}'")
 
connection = create_db_connection("localhost", "marwaan", "Marwaank17","item_details")

arrNames = []
arrAmounts = []
arrCosts = []

query1 = """
        SELECT Name
        FROM chips
        """
x = execute_get_query(connection, query1)
for i in x:
    arrNames.append(i[0])

query1 = """
        SELECT Amount
        FROM chips
        """

y = execute_get_query(connection, query1)
for i in y:
    arrAmounts.append(i[0])
    
query1 = """
        SELECT Cost
        FROM chips
        """

z = execute_get_query(connection, query1)
for i in z:
    arrCosts.append(float(i[0]))
    
print(arrNames)
print(arrAmounts)
print(arrCosts)

plt.figure(figsize=(15,5))
plt.title("Amount of items recorded")
plt.xlabel("Items")
plt.ylabel("Amount")
plt.bar(arrNames, arrAmounts)
plt.show()

plt.figure(figsize=(15,5))
plt.title("Cost of items recorded")
plt.xlabel("Items")
plt.ylabel("Cost in R")
plt.bar(arrNames, arrCosts)
plt.show()
```
